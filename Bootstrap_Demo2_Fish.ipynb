## In a pool, there are 1000 fishes (but we do not know this number in the beginning)
## How can we esitimate this number?

```python
import numpy as np
fishes = np.zeros(1000,)
print('Each 0 represents a fish!')
print(fishes)
```

##  Step 1: We get 100 fishes and mark them as 1, then put them back

```python
index = np.random.permutation(1000)[:100]
print(index)
```

```python
fishes[index]=1
print(fishes)
```

## Step 2:  get another 100 fishes and count how many of them are markes as 1?

```python
index = np.random.permutation(1000)[:100] # without replacement

count = np.sum(fishes[index])
percent = count/100
print(percent)
100/percent
```

### This means approximately 11% of the fish are marked as 1, repeat this process 100 times

```python
percents=[]
for i in range(100):
    
    index = np.random.permutation(1000)[:100]
    percent = np.sum(fishes[index])/100
    percents.append(percent)

percent_mean = np.mean(percents)
print(percent_mean)
```

## Step 3: estimate the total number

$$\dfrac{100}{x}=percent\_mean$$


$$x = \dfrac{100}{percent\_mean}$$

```python
x = np.round(100/percent_mean)
print(x)
```
